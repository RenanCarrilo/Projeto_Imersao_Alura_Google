```python
!pip install PyPDF2
!pip install -U -q google-generativeai

from PyPDF2 import PdfReader
import google.generativeai as genai
from google.colab import userdata

chave = userdata.get('chave')
genai.configure(api_key=chave)

# Configurando o Modelo
generation_config = {
  "temperature": 0,
  "top_p": 0.95,
  "top_k": 0,
  "max_output_tokens": 8192,
}

safety_settings = [
  {
    "category": "HARM_CATEGORY_HARASSMENT",
    "threshold": "BLOCK_MEDIUM_AND_ABOVE"
  },
  {
    "category": "HARM_CATEGORY_HATE_SPEECH",
    "threshold": "BLOCK_MEDIUM_AND_ABOVE"
  },
  {
    "category": "HARM_CATEGORY_SEXUALLY_EXPLICIT",
    "threshold": "BLOCK_MEDIUM_AND_ABOVE"
  },
  {
    "category": "HARM_CATEGORY_DANGEROUS_CONTENT",
    "threshold": "BLOCK_MEDIUM_AND_ABOVE"
  },
]

model = genai.GenerativeModel(model_name="gemini-1.5-pro-latest",
                              generation_config=generation_config,
                              safety_settings=safety_settings)

print("")

text = []
teste = True
teste2 = True
combined_content = ""
chat = model.start_chat ( history = [] )
pdf_base = input("Digite aqui o nome do seu arquivo pdf base:")+".pdf"

# Abrindo um arquivo PDF existente
while teste == True:
    try:
        with open(pdf_base, "rb") as input_pdf:
            # Criando um objeto PdfReader
            pdf_reader = PdfReader(input_pdf)

            # Obtendo o número de páginas do arquivo PDF
            num_pages = len(pdf_reader.pages)

            # Lendo o texto de cada página
            for page_number in range(num_pages):
                page = pdf_reader.pages[page_number]
                combined_content += page.extract_text()
                teste = False
            text.append(combined_content)
    except:
        pdf_base = input("Nome do pdf foi inválido, digite novamente (não precisa colocar a extenção .pdf no nome)")

numero_de_pdf_alunos = input("Digite quantos pdfs de alunos serão lidos:")
numero_int = int(numero_de_pdf_alunos)
n = 1

while n <= numero_int:
    combined_content = ""
    while teste2 == True:
        try:
            pdf_aluno = input("Digite o nome do aluno como está escrito no nome do PDF:")
            with open(pdf_aluno+".pdf", "rb") as input_pdf:
                # Criando um objeto PdfReader
                pdf_reader = PdfReader(input_pdf)

                # Obtendo o número de páginas do arquivo PDF
                num_pages = len(pdf_reader.pages)

                # Lendo o texto de cada página
                for page_number in range(num_pages):
                    page = pdf_reader.pages[page_number]
                    combined_content += page.extract_text()
                    teste2 = False
                text.append(combined_content)
                prompt = f"de uma nota, de forma objetiva e sem comentarios, para cada questão e uma notal final, de 0 a 10, para o aluno do {text[n]}, usando as respostas do {text[0]} como única fonte de comparação"
                response = chat.send_message(prompt)
                n = n+1
                print("Nome do aluno: "+pdf_aluno)
                print("Avaliação",response.text)
        except:
            numero_de_pdf_alunos = input("Nome do pdf de aluno foi inválido, o exemplo de nome para este arquivo é 'aluno1.pdf,aluno2.pdf...':")
            print("Altere os nomes dos arquivos e inicie o programa novamente")
```
